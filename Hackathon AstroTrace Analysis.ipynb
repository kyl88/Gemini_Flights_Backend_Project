Check for outliers in the dataset using the boxplot

```python
import seaborn as sns
sns.boxplot(df["Animal Feed"])

# Choose the maximium quantile to fill in the outliers
feed_col = df["Animal Feed"].quantile(0.995)
feed_col

df[df["Animal Feed"]> feed_col]

# the outlier is the Pig Meat
```

Data visualization

```python
# Plotting the foods with the highest total emissions
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
df_sorted = df.sort_values('Total_emissions', ascending=False)
sns.barplot(x='Total_emissions', y='Food product', data=df_sorted)
plt.title('Total Greenhouse Gas Emissions by Food Product')
plt.xlabel('Total Emissions (kgCO₂eq per kg product)')
plt.ylabel('Food Product')
plt.show()
```

```python
# Plotting the foods that transported the most - Cane suger
plt.figure(figsize=(10, 8))
df_sorted = df.sort_values('Transport', ascending=False)
sns.barplot(x='Transport', y='Food product', data=df_sorted)
plt.title('Foods that were transported the most')
plt.xlabel('Goods transported')
plt.ylabel('Food Product')
plt.show()
```

```python
# Foods that were transported the least

plt.figure(figsize=(10,8))
df.sorted = df.sort_values('Transport', ascending=True)
sns.barplot(x = 'Transport', y ='Food product', data=df.sorted)
plt.title('Foods that were transported the least')
plt.xlabel('Goods transported')
plt.ylabel('Food product')
plt.show()

```

```python
# Foods products that were packed the most

plt.figure(figsize = (10, 8))
df.sorted = df.sort_values('Packaging', ascending = False)
sns.barplot(x = 'Packaging', y = 'Food product', data=df.sorted)
plt.title('Foods that were packed the most')
plt.xlabel('Packaging')
plt.ylabel(' Food products')
plt.show()
```

```python
# When food went sent to retail stores the most

plt.figure(figsize = (10, 8))
df.sorted = df.sort_values('Retail', ascending = False)
sns.barplot(x = 'Retail', y ='Food product', data = df.sorted)
plt.title('Food products that were sent the most to retail stores')
plt.xlabel('Retail stores')
plt.ylabel('Food products')
plt.show()
```

```python
# Which foods products were produced the most in a farm

plt.figure(figsize=(10, 8))
df.sorted = df.sort_values('Farm', ascending = False)
sns.barplot(x = 'Farm', y ='Food product', data = df.sorted)
plt.title('Foods that were produced the most in a Farm')
plt.xlabel('Farm goods')
plt.ylabel('Food products')
plt.show()
```

```python
# Food products that were processed the most

plt.figure(figsize=(10, 8))
df.sorted = df.sort_values('Processing', ascending = False)
sns.barplot(x = 'Processing', y = 'Food product', data = df.sorted)
plt.title('Foods that were processed the most')
plt.xlabel('Food processed')
plt.ylabel('Food products')
plt.show()
```
